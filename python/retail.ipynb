```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/DELL/Downloads/retail_store_sales.csv")
df.head()
```

```python
df.columns.tolist()
```

```python
df.fillna(0, inplace=True)  # This modifies df in place
df.isnull().sum()
```

```python
df.info()
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
df.head()
```

```python
df['Discount Applied'] = df['Discount Applied'].replace(0, False)
df['Discount Applied'] = df['Discount Applied'].astype(bool)
df.head()


```

```python
df.isnull().sum()
```

```python
df['Item'].info()
```

```python
df.head()
```

```python
item_count = df['Item'].value_counts().reset_index()
item_count.columns = ['Item', 'Count']
item_count['Item'] =item_count['Item'].str.split('_').str[-1]
item_count =item_count.dropna()
item_count.head()

```

```python

plt.figure(figsize=(12,5))
sns.barplot(data =  item_count, x ='Count', y = 'Item', hue ='Item')
plt.show()
```

```python
cat_item = df['Category'].value_counts().reset_index()
cat_item.columns = ['Category', 'Count']
cat_item.head()
```

```python
plt.Figure(figsize=(12, 5))
sns.barplot(data = cat_item, x ='Count', y = 'Category', hue= 'Category')
plt.show()
```
