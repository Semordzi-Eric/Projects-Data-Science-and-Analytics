```python
import pandas as pd

df = pd.read_csv("C:/Users/DELL/Downloads/retail_store_sales.csv")
df.head()
```

```python
new_df =df[['Category', 'Price Per Unit','Quantity','Total Spent','Payment Method','Transaction Date','Location']]
new_df.head()
```

```python
# chaecking for missing values
import missingno as msno
import matplotlib.pyplot as plt
msno.matrix(new_df)
plt.show()


```

```python
# seperating missing values from the data
miss_df = new_df[new_df.isna().any(axis = 1)]
miss_df.head()
miss_df.describe()
```

```python
import seaborn as sns
data = new_df.dropna()
for col in data:
    plt.figure(figsize=(15,5))
    sns.histplot(df[col], bins =10)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
    
```

```python
data = new_df.dropna()
for col in data:
    plt.figure(figsize=(15,5))
    sns.histplot(data =data, x =data[col])
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
data = new_df.dropna()
for col in data:
    plt.figure(figsize=(15,5))
    sns.countplot(data =data, x =data[col])
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
data = new_df.dropna()
for col in data:
    plt.figure(figsize=(15,5))
    sns.boxplot(data =data, x= data[col],palette='coolwarm')
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```
