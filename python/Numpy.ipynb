```python
"""In NumPy, axis refers to the direction along which operations are performed on an array.

Axis 0 (rows, vertical direction) → Works column-wise (top to bottom).
Axis 1 (columns, horizontal direction) → Works row-wise (left to right)."""

cust = np.array([
    
    [0,5,2], 
    [0,2,0],
    [1,1,2],
    [2,2,1],
    [0,0,0]
])
cust.sum(axis=1)
#min val
cust.min(axis=1)
#slicing
cust[2:4]
cust[1:3,2]
cust[:,0]

```
